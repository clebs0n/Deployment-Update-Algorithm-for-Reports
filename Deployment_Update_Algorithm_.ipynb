```python
import pandas as pd
import numpy as np

logs_df = pd.read_excel('qualys.xlsx', header=None)
invent_df = pd.read_excel('invent.xlsx')

hostnames = []
statuses = []
infos = []

for i in range(0, len(logs_df), 3):
    hostname = logs_df.iloc[i, 0]
    status = logs_df.iloc[i, 1]
    info = logs_df.iloc[i+2, 0] if status == 'Failed' else np.nan

    hostnames.append(hostname)
    statuses.append(status)
    infos.append(info)

logs_new_df = pd.DataFrame({
    'Hostname': hostnames,
    'Status': statuses,
    'Info': infos
})

logs_new_df.sort_values(by='Status', ascending=False, inplace=True)
logs_new_df.drop_duplicates(subset='Hostname', keep='first', inplace=True)
print(logs_new_df)


for i in range(len(invent_df)):
    hostname_invent = invent_df.iloc[i, 0]

    if hostname_invent in logs_new_df['Hostname'].values:
        status = logs_new_df.loc[logs_new_df['Hostname'] == hostname_invent, 'Status'].values[0]
        info = logs_new_df.loc[logs_new_df['Hostname'] == hostname_invent, 'Info'].values[0]

        invent_df.iloc[i, 1] = status  # third column is at index 2

        if status == 'Failed':
            existing_info = invent_df.iloc[i, 5]  # sixth column is at index 5
            if pd.isnull(existing_info):
                invent_df.iloc[i, 5] = info
            else:
                invent_df.iloc[i, 5] = existing_info + '\n' + info

invent_df.to_excel('updated_invent.xlsx', index=False)
```
